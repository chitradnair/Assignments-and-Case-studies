# Bar graph of Male Vs Female buyers who participated in the sales

```python
slice_carsdat= carsdat[['Buyer Gender']]
slice_carsdat
```

```python
slice_carsdat['Buyer Gender'].value_counts()
```

```python
BuyGender= slice_carsdat['Buyer Gender'].value_counts()
```

```python

plt.figure(figsize=(7,5))
slice_carsdat['Buyer Gender'].value_counts().plot(kind='bar',width= 0.4, color =['Maroon','green'])
plt.title('Count of Male Vs Female Car Buyers ', fontsize=18)
plt.xlabel('Gender',fontsize=16)
plt.ylabel('Count of Buyers',fontsize=16)
```
