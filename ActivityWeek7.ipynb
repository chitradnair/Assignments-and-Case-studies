```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Reading data file 

```python
dat=pd.read_csv('Week_7_Activity_data.csv')
```

```python
dat.head()
```

# Check for Null values

```python
dat.isna().sum()
```

# Examine the variables ‘Hardness’, ‘Sulfate’, and ‘Organic_carbon’ and
comment on your observations.

```python
d1=dat[['Hardness','Sulfate','Organic_carbon']]
cormat= d1.corr()
plt.subplots(figsize=(20,8))
sns.heatmap(cormat,annot=True,vmin=-0.2,vmax=0.8,linewidth=0.2)
```

```python
# Observation : Hardness and Sulfate show weak negative correlation, Hardeness and Organic carbon show weak positive correlation
# Finally Sulfate and Organic Carbon show weak positive correlation.
# The three variables discussed are hence observed to be weakly associated to one another.
```
