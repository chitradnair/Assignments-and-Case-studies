# 4.What do you think about the scoring variation for math, reading and writing
based on
●Gender
●Test preparation course

```python
plt.figure(figsize=(12,8))
plt.subplot(2,1,1)
sns.histplot(data=stud_data, x="math score", hue="gender",multiple= 'stack',kde=True)
plt.title('Genderwise Variation of Math score',fontsize=16)
plt.show()
plt.figure(figsize=(12,8))
plt.subplot(2,1,2)
sns.histplot(data=stud_data, x="math score", hue="test preparation course",multiple='stack',kde=True)
plt.title('Test preparation Course & Variation of Math score',fontsize=16)
plt.tight_layout()
```

```python
plt.figure(figsize=(12,8))
plt.subplot(2,1,1)
sns.histplot(data=stud_data, x="reading score", hue="gender",multiple= 'stack',kde=True)
plt.title('Genderwise Variation of Reading score',fontsize=16)
plt.show()
plt.figure(figsize=(12,8))
plt.subplot(2,1,2)
sns.histplot(data=stud_data, x="reading score", hue="test preparation course",multiple='stack',kde=True)
plt.title('Test preparation Course & Variation of Reading score',fontsize=16)
plt.tight_layout()
```

```python
plt.figure(figsize=(12,8))
plt.subplot(2,1,1)
sns.histplot(data=stud_data, x="writing score", hue="gender",multiple= 'stack',kde=True)
plt.title('Genderwise Variation of Writing score',fontsize=16)
plt.show()
plt.figure(figsize=(12,8))
plt.subplot(2,1,2)
sns.histplot(data=stud_data, x="writing score", hue="test preparation course",multiple='stack',kde=True)
plt.title('Test preparation Course & Variation of Writing score',fontsize=16)
plt.tight_layout()
```
