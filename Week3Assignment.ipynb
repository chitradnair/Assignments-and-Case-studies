```python
#Import all necessary libraries in the program
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Reading dataset to python environment

```python
work_data=pd.read_csv('iris.csv')
```

```python
work_data.head()
```

# Displaying columns in the data set iris

```python
work_data.columns
```

# Calculating the mean of each column of the dataset

```python
work_data.mean()
```

# Checking null values present in data set

```python
work_data.isnull()
```

# Visualisation of the data set iris

# 1) Scatter Plot of SL vs SW

```python
plt.figure(figsize= (8,6))
plt.scatter(work_data['SL'],work_data['SW'],s=16,c='Red', marker ='^')
plt.title('Plot of SL Vs SW', fontsize=18)
plt.xlabel('SL')
plt.ylabel('SW')
```

# 2) Pair plot of all numeric columns in iris

```python
sns.pairplot(work_data)
```

# 3)Box plot of PL

```python
plt.boxplot(work_data['PL'])
plt.title('Box plot of PL')
plt.xlabel('PL')
```

# 4) Boxplot using Seaborn PL Vs PW

```python

sns.boxplot(x='PL',y='PW',data= work_data)
```
